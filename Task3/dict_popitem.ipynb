## popitem
**Remove only last key_value pair**

**Does not take position**

```python
Stock_Market= {
    "symbol": "RELIANCE", 
    "price": 2800, 
    "exchange": ["NSE", "BSE"], 
    "is_bullish": True, 
    "day_range": {"high": 2810, "low": 2790}
    }
print(f"Dictionary : {Stock_Market}")
print("--"*30)
print(f"Removed value is :{Stock_Market.popitem()}")
```
