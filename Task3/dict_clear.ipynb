## clear()
**Clears complete dictionary with only keeping empty shell of dictionary**

**clears all elements of dictionary**

```python
d1 = {"username":"Sayali",
      "email":"sayali@gmail.com",
      "age": 26,
      "Place":"Kulu-Manali",
      "Hobbie":["Dance","Reading","Sports","Music"],
      "DOB":"3/03/2002"}

print(f"Dictionary : {d1}")
print("--"*30)
d1.clear()
print(f"Dictionary after clearing: {d1}")
# Why not print(f"Dictionary after clearing: {d1.clear()}")
```
