## pop()
**Removes the element with specified key**

**If Key is not specified then removes last value**

```python
employee_detail = {
    "Id": "34853",
    "Name": "Sameer Barde",
    "email": ["sam.bar@yahoo.com","bardesameer@gamil.com"],
    "Department":"Analysis",
    "Role": "Quality Analysis",
    "Team volume": 8,
    "Manager":"Parul Bera",
    "TeamMates":["Sahil Kar","Pratham Hojiya","Renuka Rane","Tina sandel","Nihal Gurung"],
    "Floor": 4,
    "Shift":{"Night_shift":4.30, "Day_shift": 10.30},
    "Comapany": "Eclerx",
    "Size_company": 300,
    "Address": "Hinjewadi, Pune",
    "PinCode": 45444
}


print(f"Dictionary : {employee_detail}")
print("--"*30)
print(f"Removed value is : {employee_detail.pop("Team volume")}")
```
