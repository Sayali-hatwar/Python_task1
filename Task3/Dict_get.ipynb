## get()
**Return the value of specified key**

```python
a = {
    "ID": "1001",
    "order_id": "VX002",
    "sales_id": ["TH2304","FG3957938","DF345302"],
    "Product": {"Furniture":["Chair","Table","Stool","Bed"], 
                "Kitche":["Knife","Bowl","Chopping Slate","Pot"],
                "Electric":["Bulb","Fan","TV"]},
    "Shop" : "IKEA"
}

print(f"Dictionary : {a}")
a1 = a.get("ID")
print(f"We get the Value of the key 'ID' : {a1}")
```
