## setdefault()
**Returns the value of the specified key. If the key does not exist: insert the key, with the specified value**

**If key does not exist then give default value to be filled if method does not get anything. It can be anything**

```python
Airline_Ticket = {
    "pnr": "XY99", 
    "from": "BOM", 
    "stops": ["DXB", "LHR"], 
    "airline": "Emirates", 
    "seat": {"row": 12, "pos": "Window"}
    }

print(f"Dictionary : {Airline_Ticket}")
print("--"*20)
at = Airline_Ticket.setdefault("pnr")
print(f"Value after setting default : {at}")
```

```python
Music_Album = {
    "title": "Evermore", 
    "artist": "Taylor Swift", 
    "tracks": ["Willow", "Gold Rush"], 
    "year": 2020, 
    "label": {"name": "Republic"}
    }
print(f"Dictionary : {Music_Album}")
print("--"*20)
gt = Music_Album.setdefault("pnr","None")  ## putting key outside of dict gives default as output 
print(f"Value after setting default : {gt}")
```
