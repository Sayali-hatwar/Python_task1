# **String Methods**

## **1. UpperCase**

```python
d1 = "I am a Human"
print(f"Upper string text: {d1.upper()}")

d2 = "We live on Earth"
print(f"Upper string text: {d2.upper()}")

d3 = "This is a tiger"
print(f"Upper string text: {d3.upper()}")

d4 = "Run a jupiter notebook"
print(f"Upper string text: {d4.upper()}")

d5 = "My pet is a cat"
print(f"Upper string text: {d5.upper()}")
```

## **2. LowerCase**

```python
l1 = "I do Practice"
print(f"Lower string text: {l1.lower()}")

l2 = "I learn Python"
print(f"Lower string text: {l2.lower()}")

l3 = "I make Project"
print(f"Lower string text: {l3.lower()}")

l4 = "This is my house"
print(f"Lower string text: {l4.lower()}")

l5 = "I am going to Party"
print(f"Lower string text: {l5.lower()}")
```

## **3. Capitalize**

```python
c1 = "Today is Monday. It is hot today."
print(f"Capitalize string text: {c1.capitalize()}")

c2 = "I am Travelling. Are you coming with me?"
print(f"Capitalize string text: {c2.capitalize()}")

c3 = "I use JUPITER Notebook, rather than VS code."
print(f"Capitalize string text: {c3.capitalize()}")

c4 = "What you do for livign?"
print(f"Capitalize string text: {c4.capitalize()}")

c5 = "Is everything ALRIGHT?"
print(f"Capitalize string text: {c5.capitalize()}")
```

## **4. Title**

```python
t1 = "I am drinking water"
print(f"Title string text: {t1.title()}")

t2 = "are you eating Apple? it's good."
print(f"Title string text: {t2.title()}")

t3 = "We are going to dubai. Do you want to join us?"
print(f"Title string text: {t3.title()}")

t4 = "I use Laptop for my work"
print(f"Title string text: {t4.title()}")

t5 = "I have big teddy"
print(f"Title string text: {t5.title()}")
```

## **5. Strip** -- only strips starting and ending unnecessary spaces

```python
s1 = "     I  am wearing Long dress"
print(f"Strip string text: {s1.strip()}")

s2 = "Are you writting   in notebook        "
print(f"Strip string text: {s2.strip()}")

s3 = "I am going to meet my friend        "
print(f"Strip string text: {s3.strip()}")

s4 = "  This is beautiful"
print(f"Strip string text: {s4.strip()}")

s5 = "Where are      you from"
print(f"Strip string text: {s5.strip()}")
```

## **6. Right Strip and Left Strip**

```python
r1 = "I love my family      "
print(f"Right Strip string text: {r1.rstrip()}")

r2 = "Today is Tuesday      "
print(f"Right Strip string text: {r2.rstrip()}")

r3 = "NExt month is June      "
print(f"Right Strip string text: {r3.rstrip()}")

l1 = "     We are           eco-friendly"
print(f"Left Strip string text: {l1.lstrip()}")

l2 = "        I am going on trip."
print(f"Left Strip string text: {l2.lstrip()}")

l3 = "      This    is Preety"
print(f"Left Strip string text: {l3.lstrip()}")
```

## **7. Replace**

```python
q1 = "I am going to my house"
print(f"Replace string text: {q1.replace('house', 'apartment')}")

q2 = "This is my hostel"
print(f"Replace string text: {q2.replace('hostel','institute')}")

q3 = "I am working"
print(f"Replace string text: {q3.replace('working','dancing')}")

q4 = "This is my bag"
print(f"Replace string text: {q4.replace('is','is not')}")

q5 = "I am going to party"
print(f"Replace string text: {q5.replace('party','school')}")
```
